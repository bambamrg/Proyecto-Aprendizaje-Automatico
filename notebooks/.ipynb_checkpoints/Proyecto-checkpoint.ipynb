```python
# Se importan las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Se cargan los datos desde un archivo excel
df = pd.read_excel("../data/raw/SAT-SS-BU_2017-2022.xlsx")
```

```python
# Se muestran las primeras 5 filas
df.head()
```

```python
# Realizamos una exploración gráfica con un boxplot de las variables numéricas
# Para crear un boxplot de todas las columnas numéricas:

plt.boxplot(df.select_dtypes(include=['float64', 'int64']))
plt.title('Boxplots de variables numéricas')
plt.show()
```

```python
# Crear un boxplot horizontal con etiquetas
sns.boxplot(data=df, orient='h')
plt.title('Distribución de variables numéricas')
plt.xlabel('Valor')
plt.ylabel('Variable')
plt.show()
```
